```python
from sklearn import tree
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score, log_loss
```

```python
titanic = pd.read_csv('https://stepik.org/media/attachments/course/4852/titanic.csv')
```

```python
X=titanic.drop(['Name','PassengerId','Survived','Ticket','Cabin'],axis=1)
y=titanic.Survived
X=pd.get_dummies(X)  # Заменяем текст на True False
X=X.fillna({'Age' : X.Age.median()}) # заменяем пустой возраст на медиану
```

```python
clf_rf=RandomForestClassifier()
```

```python
parametrs={'n_estimators':[10,20,30,40],'max_depth':[3,5,7,10,12,15]}  # задаем параметры вручную промежутками
```

```python
grid_search_cv=GridSearchCV(clf_rf,parametrs,cv=5) #GridSearchCV помогает автоматизировать процесс поиска оптимальных значений гиперпараметров
```

```python
X_test, X_train, y_test, y_train = train_test_split(X,y, test_size=0.3 , random_state=42 )
```

```python
grid_search_cv.fit(X_train,y_train)
```

```python
grid_search_cv.score(X_test,y_test)
```

```python
grid_search_cv.best_params_  #Поски лучших параметров
```

```python
best_clf_rf=grid_search_cv.best_estimator_  # Лучшая модель 
```

```python
y_pred=best_clf_rf.predict(X_test)
```

```python
print(f'accuracy: {best_clf_rf.score(X_test, y_test):.3f}') 
print(f'precision: {precision_score(y_test, y_pred):.3f}') 
print(f'recall: {recall_score(y_test, y_pred):.3f}')
print(f'f1: {f1_score(y_test, y_pred):.3f}')
```

```python
feature_importatnce=best_clf_rf.feature_importances_ # оценка важности каждого признака
```

```python
feature_importatnce_df=pd.DataFrame({'features' :list(X_train),'feature_importatnce':feature_importatnce })
```
