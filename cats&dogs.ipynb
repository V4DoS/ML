```python
from sklearn import tree
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
animals = pd.read_csv('dogs_n_cats.csv')
test=pd.read_json('dataset_209691_15.txt')
```

```python
filtered_animals = animals[animals['Высота'] >= 11.8]
count = filtered_animals.shape[0]
count
```

```python
X=animals.drop(['Вид'],axis=1)
y=animals.Вид
```

```python
clf=tree.DecisionTreeClassifier(criterion='entropy') #Энтропия
```

```python
clf.fit(X,y) # обучение
```

```python
plt.figure(figsize=(30, 20))

# Отрисовка дерева решений
tree.plot_tree(clf, filled=True, rounded=True)

# Установка заголовка
plt.title('Дерево решений с критерием энтропия')

# Сохранение графика в файл перед его показом
plt.savefig('decision_tree_entropy.png', dpi=300)
```
